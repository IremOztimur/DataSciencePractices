## Understanding the Concept of Basis

In technical terms,the basis of a vector space is a set of linearly independent vectors that span the full space.

To demystify the concept of vector basis, consider the following analogy:
    
Imagine you want to build a house, and you have a set of tools that includes a hammer, a saw, and a screwdriver. Each tool serves a specific purpose in building different parts of the house.

Now, let's say these tools are like basis vectors. The hammer, saw, and screwdriver form a basis for your set of tools because, with just these three tools, you can construct the entire house. You can use the hammer for nails, the saw for cutting wood, and the screwdriver for assembling parts.

In this analogy:

- The tools (hammer, saw, screwdriver) are like basis vectors.
- Building the house is like creating any vector in the space.

With this basis, you can represent any construction task using combinations of the hammer, saw, and screwdriver. Similarly, in linear algebra, a basis is a set of vectors that can be combined to represent any vector in a given vector space.

If you had more tools (additional basis vectors), they might not be necessary for basic construction, but they could offer more flexibility or represent additional ways of building the house. However, as long as you have the essential tools (basis vectors), you can still accomplish the task.

# Visualizing Matrix Multiplication

$\hspace{1cm}\textit{transformed } \vec{v} = x\mathbin{\color{green}{\hat{i_T}}} +\, y\, \mathbin{\color{red}{\hat{j_T}}} $

$\hspace{2.3cm}$*Equation 1*


This method of transforming a vector through use of the *transformed* basis vectors is the same as matrix multiplication (see *Equation 2*).

$\hspace{1cm} \begin{bmatrix} \mathbin{\color{green}a} & \mathbin{\color{red}b}\\ \mathbin{\color{green}c} & \mathbin{\color{red}d} \end{bmatrix} \begin{bmatrix} x\\ y\end{bmatrix} = x \begin{bmatrix}\mathbin{\color{green}a}\\ \mathbin{\color{green}c} \end{bmatrix} + y \begin{bmatrix} \mathbin{\color{red}b}\\ \mathbin{\color{red}d} \end{bmatrix} = \begin{bmatrix} \mathbin{\color{green}a}x + \mathbin{\color{red}b}y\\ \mathbin{\color{green}c}x + \mathbin{\color{red}d}y\end{bmatrix}$ 

$\hspace{4.1cm}$*Equation 2*



## Graphing a Vector $\vec{v}$ Decomposed into Basis Vectors $\vec{\hat{i}}$ and $\vec{\hat{j}}$

$\hspace{1cm}\vec{v} = \begin{bmatrix} -1\\ 2\end{bmatrix}$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

v = np.array([-1,2])

i_hat = np.array([1,0])

j_hat = np.array([0,1])

v_ihat = v[0] * i_hat

v_jhat = v[1] * j_hat

ax = plt.axes()

ax.plot(0,0,'or')

ax.arrow(0, 0, *v_ihat, color='g', linestyle='dotted', linewidth=2.5, head_width=0.30,
         head_length=0.35)

ax.arrow(v_ihat[0], v_ihat[1], *v_jhat, color='r', linestyle='dotted', linewidth=2.5,
         head_width=0.30, head_length=0.35)

ax.arrow(0, 0, *v, color='b', linewidth=2.5, head_width=0.30, head_length=0.35)


plt.xlim(-4, 2)

major_xticks = np.arange(-4, 2)
ax.set_xticks(major_xticks)


plt.ylim(-2, 4)

major_yticks = np.arange(-2, 4)
ax.set_yticks(major_yticks)

plt.grid(b=True, which='major')

plt.show()
```

## *Transforming* a Vector using *Transformed* Vectors $\vec{\hat{i_T}}$ and $\vec{\hat{j_T}}$

$\hspace{1cm}\vec{v} = \begin{bmatrix} -1\\ 2\end{bmatrix}$

$\hspace{1cm}\vec{\mathbin{\color{green}{\hat{i_T}}}} = \begin{bmatrix}\mathbin{\color{green}3}\\ \mathbin{\color{green}1} \end{bmatrix}$

$\hspace{1cm}\vec{\mathbin{\color{red}{\hat{j_T}}}} = \begin{bmatrix}\mathbin{\color{red}1}\\ \mathbin{\color{red}2} \end{bmatrix}$


```python
v = np.array([-1, 2])

ihat_t = np.array([3, 1])

jhat_t = np.array([1, 2])

v_ihat_t = v[0] * ihat_t

v_jhat_t = v[1] * jhat_t

v_t = v_ihat_t + v_jhat_t

ax = plt.axes()

ax.plot(0,0,'or')

ax.arrow(0, 0, *v_ihat_t, color='g', linestyle='dotted', linewidth=2.5, head_width=0.30,
         head_length=0.35)

ax.arrow(v_ihat_t[0], v_ihat_t[1], *v_jhat_t, color='r', linestyle='dotted', linewidth=2.5,
         head_width=0.30, head_length=0.35)

ax.arrow(0, 0, *v, color='skyblue', linewidth=2.5, head_width=0.30, head_length=0.35)

ax.arrow(0, 0, *v_t, color='b', linewidth=2.5, head_width=0.30, head_length=0.35)


plt.xlim(-4, 2)

major_xticks = np.arange(-4, 2)
ax.set_xticks(major_xticks)


plt.ylim(-2, 4)

major_yticks = np.arange(-2, 4)
ax.set_yticks(major_yticks)

plt.grid(b=True, which='major')

plt.show()
```

## Matrix Multiplication
It is the same vector transformation from the section above can be achieved with matrix multiplication (*Equation 2*).
   
$\hspace{1cm}\vec{v} = \begin{bmatrix} -1\\ 2\end{bmatrix}$

$\hspace{1cm}\vec{ij} = \begin{bmatrix} \mathbin{\color{green}3} & \mathbin{\color{red}1}\\ \mathbin{\color{green}1} & \mathbin{\color{red}2}\end{bmatrix}$

```python
v = np.array([-1,2])

ij = np.array([[3, 1],[1, 2]])

v_t = np.matmul(v, ij)

print("\nMatrix ij:", ij, "\nVector v:", v, "\nTransformed Vector v_t:", v_t, sep="\n")
```
